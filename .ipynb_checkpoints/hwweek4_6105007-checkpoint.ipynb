```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
from datetime import datetime
```

```python
data = pd.read_csv('COVID19.csv')
#data
```

```python
th = np.where(data == 'THA')
#th
#th[0]
#th[1]
caseth = np.array(data['cases'][np.min(th[0]):np.max(th[0])+1])
#caseth
```

```python
print('COVID-19 in Thailand')
print('Cases :',np.sum(caseth))
print("Percentage of infected people in the entire population :", float("{:.4f}".format(np.sum(caseth)/69428524*100)),'%')
print('Deaths :', np.sum(data['deaths'][np.min(th[0]):np.max(th[0])]))
print('Percentage of deat per case :', float("{:.4f}".format(np.sum(data['deaths'][np.min(th[0]):np.max(th[0])])/np.sum(caseth)*100)),'%')
```

```python
def dataplot(a):
    us = np.array(np.where(data == a))
    case = np.array(data['cases'][np.min(us[0]):np.max(us[0])+1])
    days = data['dateRep'][np.min(us[0]):np.max(us[0])+1]#days
    day = [datetime.strptime(x,'%d/%m/%Y' ) for x in days]

    sum = [np.sum(case),]
    s = np.sum(case)
    for i in range(len(days)-1):
        s = s-case[i]
        sum = np.append(sum,s)
    return day,sum

#dataplot('USA')
```

```python
figure(num=None, figsize=(10, 10), dpi=500, facecolor='w', edgecolor='k')
a,b = dataplot('USA')
plt.plot_date(a,b,'*', label='USA')

a,b = dataplot('THA')
plt.plot_date(a,b,'x', label='Thailand')

a,b = dataplot('CHN')
plt.plot_date(a,b,'.', label='China')

a,b = dataplot('GBR')
plt.plot_date(a,b,'2', label='UK')

a,b = dataplot('TWN')
plt.plot_date(a,b,'3', label='Taiwan')

a,b = dataplot('SWE')
plt.plot_date(a,b,'+', label='Sweeden')

a,b = dataplot('ESP')
plt.plot_date(a,b,'4', label='Spain')

a,b = dataplot('KOR')
plt.plot_date(a,b,'>k', label='Koria')

a,b = dataplot('SGP')
plt.plot_date(a,b,'<r', label='Singapor')

a,b = dataplot('RUS')
plt.plot_date(a,b,'^', label='Rusia')

a,b = dataplot('PRT')
plt.plot_date(a,b,'.g', label='Portugal')

a,b = dataplot('NOR')
plt.plot_date(a,b,'8', label='Norway')

a,b = dataplot('PHL')
plt.plot_date(a,b,'X', label='Phillipines')

a,b = dataplot('JPN')
plt.plot_date(a,b,'o', label='Japan')

plt.yscale('log')

plt.xticks(rotation=45,fontsize=16)
plt.yticks(fontsize=16)
plt.legend(loc='center left',fontsize=16, bbox_to_anchor=(1, 0.5))
plt.title('COVIS-19 cases per population \n from 31-12-2019 to 12-04-2020',fontsize=20, fontweight='bold')
plt.xlabel('Day',fontsize=20, fontweight='bold')
plt.ylabel('Cases',fontsize=20, fontweight='bold')
plt.grid(True, which="both", ls="--",alpha=0.5)
plt.show()

#dataplot('RUS')[1]
```

```python
def dataplot2(a):
    us = np.array(np.where(data == a))
    case = np.sum(np.array(data['cases'][np.min(us[0]):np.max(us[0])+1]))
    death = np.sum(np.array(data['deaths'][np.min(us[0]):np.max(us[0])+1]))
    popu = data['popData2018'][np.min(us[0])]

    cpp = case/popu*100
    dpc = death/case*100

    popu = np.sqrt(popu)/50
    
    return cpp,dpc,popu
```

```python
figure(num=None, figsize=(10, 10), dpi=500, facecolor='w', edgecolor='k')
plt.set_cmap('hot')
a,b,c = dataplot2('USA')
plt.scatter(a,b,c, label='USA',alpha=0.5)

a,b,c = dataplot2('THA')
plt.scatter(a,b,c, label='Thailand',alpha=0.5)

a,b,c = dataplot2('CHN')
plt.scatter(a,b,c, label='China',alpha=0.5)

a,b,c = dataplot2('GBR')
plt.scatter(a,b,c, label='UK',alpha=0.5)

a,b,c = dataplot2('TWN')
plt.scatter(a,b,c, label='Taiwan',alpha=0.5)

a,b,c = dataplot2('SWE')
plt.scatter(a,b,c, label='Sweeden',alpha=0.5)

a,b,c = dataplot2('ESP')
plt.scatter(a,b,c, label='Spain',alpha=0.5)

a,b,c = dataplot2('KOR')
plt.scatter(a,b,c,label='Koria',alpha=0.5)

a,b,c = dataplot2('SGP')
plt.scatter(a,b,c, label='Singapor',alpha=0.5)

a,b,c = dataplot2('RUS')
plt.scatter(a,b,c, label='Rusia',alpha=0.5)

a,b,c = dataplot2('PRT')
plt.scatter(a,b,c, 'lime',label='Portugal',alpha=0.5)

a,b,c = dataplot2('NOR')
plt.scatter(a,b,c,'purple', label='Norway',alpha=0.5)

a,b,c = dataplot2('PHL')
plt.scatter(a,b,c,'slateblue', label='Phillipines',alpha=0.5)

a,b,c = dataplot2('JPN')
plt.scatter(a,b,c,'slategrey', label='Japan',alpha=0.5)

plt.legend(loc='center left',fontsize=16, bbox_to_anchor=(1, 0.5))
plt.title('COVIS-19 cases and deaths \n from 31-12-2019 to 12-04-2020',fontsize=20, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Cases per population (%)',fontsize=20, fontweight='bold')
plt.ylabel('Death per cases (%)',fontsize=20, fontweight='bold')
plt.grid(True, which="both", ls="--")
# plt.xscale('log')
# plt.yscale('log')
plt.show()
```
