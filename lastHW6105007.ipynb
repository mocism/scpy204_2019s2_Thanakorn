```python
from astroquery.sdss import SDSS
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
data = SDSS.query_sql("select                       s.z, s.zErr, p.petroMag_g, p.petroMagErr_g, s.bestObjID                      from                        photoObj p                      join                        specObj s ON s.bestObjID = p.objID                        where                        s.class = 'qso' and s.z between 0.3 and 0.4 and s.zWarning = 0 and p.petroMagErr_g	between 0 and 1 and s.zErr between 0 and 0.01")
data
```

```python
z = np.array(data['z'])
zerr = np.array(data['zErr'])
mag = np.array(data['petroMag_g'])
magerr = np.array(data['petroMagErr_g'])
```

```python
figure(num=None, figsize=(10, 10), dpi=500, facecolor='w', edgecolor='k')

plt.errorbar(z,mag,magerr,zerr,'.')

plt.ylim(10,30)
plt.xlim(0.29,0.41)
plt.title('Graph between Redshift and Petrosiam magnitude',fontsize=20, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Redshift',fontsize=20, fontweight='bold')
plt.ylabel('Petrosian Mag. (mag) ',fontsize=20, fontweight='bold')
plt.show()
```
